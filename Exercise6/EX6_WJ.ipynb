```python
import geopandas as gpd
import matplotlib.pyplot as plt
import folium
```

```python
town_halls = gpd.read_file("Town_Halls_gdb_-4946773943493791132/TOWNHALLS_PT_MEMA.shp")

town_halls.head()
```

```python
town_halls.crs
```

```python
mass_towns = gpd.read_file("mass_towns.geojson")

mass_towns.head()
```

```python
mass_towns.crs
```

```python
town_halls = town_halls.to_crs("EPSG:4326")
mass_towns = mass_towns.to_crs("EPSG:4326")
```

```python
ax = mass_towns.plot(
    color="lightgray",
    edgecolor="black",
    figsize=(12, 10)
)

town_halls.plot(
    ax=ax,
    color="red",
    markersize=10
)

plt.title("Massachusetts Town Halls")
plt.show()
```

```python
m = folium.Map(
    location=[42.3, -71.8],
    zoom_start=8
)
```

```python
folium.GeoJson(
    mass_towns,
    name="Massachusetts Town Boundaries"
).add_to(m)
```

```python
folium.GeoJson(
    town_halls,
    name="Town Halls"
).add_to(m)
```

```python
census_data = gpd.read_file("CENSUS2020TOWNS_SHP/CENSUS2020TOWNS_POLY.shp")
census_data.columns
```

```python
m2 = folium.Map(
    location=[42.3, -71.8],
    zoom_start=8
)

folium.Choropleth(
    geo_data=census_data,
    data=census_data,
    columns=["TOWN20", "POP2020"],
    key_on="feature.properties.TOWN20",
    fill_color="YlOrRd",
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name="Population 2020"
).add_to(m2)

folium.LayerControl().add_to(m2)

m2
```
