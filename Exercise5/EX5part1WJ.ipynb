```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point
```

```python
# 1. Read the CSV file into a pandas DataFrame

df = pd.read_csv('C://Users//mushr//GEOG300S26LM//Arabica_Geopy.csv') #

# Ensure your SV has columns named 'Latitude' and 'Longitude' (adjust names if needed)
```

```python
# 2. Convert Latitude and Longitude columns to a list of shapely Point objects
# Note: The order for points_from_xy is (longitude, latitude)
geometry = gpd.points_from_xy(df['longitude'], df['latitude']) #
```

```python
# 3. Create a GeoDatarame from the Dataframe and the new geometry
gdf = gpd.GeoDataFrame(df, geometry=geometry)
# 4. Set the Coordinate Reference System (CRS)
# EPSG: 4326 is the standard for latitude and longitude (WGS 84)
gdf = gdf.set_crs("EPSG:4326")#
```

```python
# If using a basemap, plot it first, then plot your points on top
# fig, ax = plt.subplots (figsize=(10, 6))
#world.plot(ax=ax, color="lightgray')
# gdf.plot(ax=ax, marker='o', color='red', markersize=5)

# Or simply plot the points
gdf.plot (marker= 'o', color='red', markersize=5, figsize=(10, 6))
plt.title("Points from CSV on a Map")
plt.show()
```
