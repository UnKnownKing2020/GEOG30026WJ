```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt

# Load CSV
df = pd.read_csv("EX4_WJ_geopy.csv")

# Create geometry points
geometry = gpd.points_from_xy(
    df["longitude"],
    df["latitude"]
)

# Create GeoDataFrame
gdf = gpd.GeoDataFrame(
    df,
    geometry=geometry
)

# Set coordinate system
gdf.crs = "EPSG:4326"

# Load country polygons
countries = gpd.read_file("countries.geojson")

# Plot countries
ax = countries.plot(
    color="lightblue",
    edgecolor="black",
    figsize=(15,10)
)

# Plot points
gdf.plot(
    ax=ax,
    color="red",
    markersize=40
)

plt.title("Motorcycle Production Countries")

plt.show()
```
